```python
# PREGUNTA 1: Probabilidad de que un estudiante obtenga menos de 72 puntos en escritura ---
print("PREGUNTA 1: P(Escritura < 72)")

N_escr_menor_72 = df[df['writing score'] < 72].shape[0]

prob_escrita_menor_72 = N_escr_menor_72 / N_total

print(f"Total de estudiantes (N): {N_total}")
print(f"Estudiantes con escritura < 72: {N_escr_menor_72}")
print(f"Probabilidad P(W < 72): {prob_escrita_menor_72:.3f} ({prob_escrita_menor_72 * 100:.1f}%)")


# --- PREGUNTA 2: Probabilidad de que un estudiante obtenga más de 80 puntos en matemáticas Y en escritura ---
print("\n--- PREGUNTA 2: P(Matemáticas > 80 Y Escritura > 80) ---")

condicion_interseccion = (df['math score'] > 80) & (df['writing score'] > 80)
N_ambos_mayor_80 = df[condicion_interseccion].shape[0]

prob_ambos_mayor_80 = N_ambos_mayor_80 / N_total

print(f"Total de estudiantes (N): {N_total}")
print(f"Estudiantes con Matemáticas > 80 Y Escritura > 80: {N_ambos_mayor_80}")
print(f"Probabilidad P(M > 80 ∩ W > 80): {prob_ambos_mayor_80:.3f} ({prob_ambos_mayor_80 * 100:.1f}%)")


# --- PREGUNTA 3: Nivel educativo de los padres con mayor probabilidad condicional P(Matemáticas > 90 | Nivel de los Padres) ---
print("\n--- PREGUNTA 3: Mayor P(Matemáticas > 90 | Nivel de los Padres) ---")

# Identificar el Evento A (Matemáticas > 90) y crear un DataFrame booleano
evento_A = df['math score'] > 90

# Obtener la lista de todos los niveles educativos de los padres (Eventos B)
niveles_padres = df['parental level of education'].unique()

probabilidades_condicionales = {}

for nivel in niveles_padres:
    evento_B = df['parental level of education'] == nivel

    # Numerador: P(A ∩ B_i) -> Número de estudiantes con M > 90 Y nivel educativo B_i
    N_A_y_B = (evento_A & evento_B).sum()

    # Denominador: P(B_i) -> Total de estudiantes con el nivel educativo B_i
    N_B = evento_B.sum()
    
    # Cálculo de la probabilidad condicional: P(A | B_i) = N(A ∩ B_i) / N(B_i)
    if N_B > 0:
        prob = N_A_y_B / N_B
        probabilidades_condicionales[nivel] = prob
    else:
        probabilidades_condicionales[nivel] = 0

series_prob = pd.Series(probabilidades_condicionales).sort_values(ascending=False)

print("\nCálculo de la probabilidad condicional por nivel educativo:")
print(series_prob.to_string(float_format="%.3f"))

mejor_nivel = series_prob.index[0]
mayor_prob = series_prob.iloc[0]

print("\nNivel educativo de los padres que ofrece la MAYOR probabilidad condicional:")
print(f"-> {mejor_nivel} con P(M > 90 | Nivel) = {mayor_prob:.3f} ({mayor_prob * 100:.1f}%)")
```
