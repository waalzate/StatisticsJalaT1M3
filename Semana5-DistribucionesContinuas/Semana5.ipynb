# Semana 5: Distribuciones continuas


La relevancia de las distribuciones continuas reside en la diversidad de fenómenos del mundo real que requieren un análisis de resultados cuyos valores pueden variar en un intervalo de números reales. Una variable aleatoria continua se refiere a aquella que puede tomar cualquier valor numérico dentro de un intervalo o colección de intervalos. 

Los resultados experimentales que se basan en mediciones de magnitudes como tiempo, peso, distancia y temperatura pueden ser descritos mediante variables aleatorias continuas
 
A pesar de la imposibilidad de presentar en una tabla la distribución de probabilidades de una variable que puede abarcar cualquier valor dentro de un rango, como la altura de las personas, es posible describirlo mediante una función llamada f(x), conocida como "función de densidad de probabilidad". Esta función proporciona la capacidad de comprender cómo las probabilidades se distribuyen en distintos valores de la variable, ofreciendo así una representación más completa y detallada de la información.

## Función de densidad

Suponga que $X$ es una variable aleatoria continua. La función de probabilidad o función de densidad de probabilidad (fdp) de $X$ se representa como $f(x)$, es una función no negativa tal que para cualquier par de números reales $a$ y $b$, donde $a \leq b$, la probabilidad de que $X$ esté en el intervalo $[a, b]$ se expresa como la integral definida de $f(x)$ en ese intervalo:

$$
P(a \leq X \leq b) = \int_{a}^{b} f(x) \, dx.
$$

Si $f(x)$ es una función de densidad entonces satisface la siguiente condición:

$$
\int_{-\infty}^{\infty} f(x) \, dx = 1.
$$
## Propiedades de una distribución continua:  

Si $X$ tiene una distribución continua se cumple que:

1. La probabilidad de que $X$ sea igual a un valor específico $a$ es cero: 
$$P(X = a) = 0$$.

2. La probabilidad de que $X$ sea menor o igual a un valor $a$ es igual a la probabilidad de que $X$ sea estrictamente menor que $a$: 

$$P(X \leq a) = P(X < a)$$.

3. La probabilidad de que $X$ sea mayor que un valor $a$ es igual a $1$ menos la probabilidad de que $X$ sea menor que $a$: 

$$P(X > a) = 1 - P(X < a)$$


## Valor esperado y Varianza

Si $X$ es una variable aleatoria con una distribución de probabilidad $f(x)$, la media o valor esperado de $X$ se define como:

$$
\mu = E(X) = \int_{-\infty}^{\infty} x f(x) \, dx
$$


La varianza de $X$ se define como:

$$
\sigma^2 = V(X) = \int_{-\infty}^{\infty} (x - \mu)^2 f(x) \, dx
$$


## Distribución uniforme

Se dice que una variable aleatoria continua $X$ tiene una distribución uniforme en el intervalo $[a, b]$ si la función de densidad de probabilidad de $X$ está dada por

$$
f(x) = \begin{cases} 
\frac{1}{b - a} & \text{si } a \leq x \leq b \\ 
0 & \text{en otro caso}
\end{cases}
$$

 
 

### Ejemplo: ( Distribución uniforme)

Supongamos que el tiempo que una persona pasa en un parque temático, desde la llegada hasta la salida, sigue una distribución uniforme continua en un intervalo de tiempo máximo de seis horas. Con frecuencia, los visitantes pueden optar por pasar diferentes cantidades de tiempo, y la duración X de la visita se asume que sigue una distribución uniforme en el intervalo $[0, 6]$.

- ¿Cuál es la función de densidad de probabilidad? 
- ¿Cuál es la probabilidad de que cualquier visitante permanezca al menos 3 horas?

### Solución:

- ¿Cuál es la función de densidad de probabilidad?

$$
f(x) = \begin{cases} 
\frac{1}{6} & \text{si } 0 \leq x \leq 6 \\ 
0 & \text{en otro caso}
\end{cases}
$$

- ¿Cuál es la probabilidad de que cualquier conferencia determinada dure al menos 3 horas?

$$
P(X \geq 3) = \int_{3}^{6} f(x) \, dx = \int_{3}^{6} \frac{1}{6} \, dx = \frac{1}{6} \int_{3}^{6} 1 \, dx = \frac{1}{6} \times (6 - 3) = \frac{3}{6} = \frac{1}{2}
$$




### Ejemplo: (Distribución uniforme)

Una variable aleatoria sigue una distribución uniforme en el intervalo de 5 a 10. Obtén la media y la varianza.

```python
 
# límite inferior de la distribución uniforme
lower_limit = 5

# límite superior de la distribución uniforme
upper_limit = 10

# calcular la media de la distribución uniforme
uniform_mean = (upper_limit + lower_limit) / 2

# imprimir la media
print('Media:', uniform_mean)

# calcular la varianza de la distribución uniforme
uniform_variance = (upper_limit - lower_limit)**2 / 12

# imprimir la varianza
print('Varianza:', uniform_variance)
```

## Distribución Normal

La distribución normal, también conocida como gaussiana, encuentra aplicación en estadística para modelar diversos fenómenos en campos como la física, química, biología y ciencias sociales. Por ejemplo, al analizar la altura de las personas, se observa que la distribución normal proporciona un ajuste preciso, permitiendo comprender su comportamiento y tomar mejores decisiones. 
 
 
Además, esta distribución es esencial en encuestas de opinión, facilitando inferencias y estimaciones basadas en sus propiedades. Al comprender la distribución normal y sus características, se mejora la toma de decisiones al seleccionar métodos estadísticos y técnicas de análisis apropiados para el manejo eficaz de los datos.


Decimos que una variable aleatoria continua $X$ tiene una distribución normal con media $\mu$ y desviación estándar $\sigma$ si su función de probabilidad $f(x)$ está dada por

$$
f(x) = \frac{1}{\sqrt{2 \pi \sigma^2}} e^{-\frac{(x - \mu)^2}{2 \sigma^2}}
$$

para todo $x$ en los reales.


En el caso de una variable aleatoria que sigue una distribución normal con media $\mu = 0$ y desviación estándar $\sigma^2 = 1$, se la denomina variable normal estándar y comúnmente se denota con la letra $Z$. La gráfica de esta distribución se muestra en la siguiente gráfica. Al sustituir $\mu = 0$ y $\sigma^2 = 1$ en la función de densidad anterior, obtenemos que:

$$
f(x) = \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}}
$$


## Gráfica de la función de densidad de una distribución normal estándar

```python
#Código en python para la gráfica

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Definir los parámetros de la distribución normal estándar
mu = 3
sigma = 2

# Crear un rango de valores x
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 1000)

# Calcular la función de densidad de probabilidad (pdf) para cada valor de x
pdf = stats.norm.pdf(x, mu, sigma)

# Crear la gráfica
plt.plot(x, pdf, color='red', label=' f(x)')

# Agregar título y etiquetas
plt.title('Distribución Normal Estándar')
plt.xlabel('x')
plt.ylabel('  y=f(x)  ')

# Mostrar leyenda
plt.legend()

# Mostrar la gráfica
plt.show()
```

## Estandarización de una distribución normal.

La estandarización de una distribución normal consiste en transformar una variable aleatoria con media $\mu$ y varianza $\sigma^2$ en una variable que sigue una distribución normal estándar, es decir, con media 0 y desviación estándar 1. Esta transformación se realiza mediante la fórmula:

$$
Z = \frac{X - \mu}{\sigma}
$$

 
Una vez obtenida $Z$, se pueden utilizar las propiedades de la distribución normal estándar para realizar cálculos y comparaciones de manera más sencilla. La estandarización es una herramienta común en estadística para normalizar y simplificar los cálculos en situaciones que involucran distribuciones normales.

Para calcular la probabilidad $P(X < x)$ de una variable aleatoria normal, se realiza la integral sobre la función de densidad de probabilidad:

$$
P(X < x) = \int_{-\infty}^{x} f(x) \, dx = \frac{1}{\sigma \sqrt{2\pi}} \int_{-\infty}^{x} \exp{\left(-\frac{(x - \mu)^2}{2\sigma^2}\right)} \, dx
$$

Debido a que esta integral no puede resolverse mediante técnicas de integración convencionales, se recurre a tablas de distribución o herramientas computacionales como Python, R, o Excel para estimarla.


## Ejemplo ( Estandarización de una variable aleatoria)

Supongamos que tenemos una variable aleatoria $X$ que representa las puntuaciones en un examen, con una media $\mu = 70$ y una desviación estándar $\sigma = 10$. Queremos encontrar la probabilidad de que un estudiante obtenga una puntuación menor que 85, es decir, $P(X<85)$. Se consideran los siguientes pasos:

1. **Transformar $X$ en $Z$**:
   Para estandarizar la variable $X$, utilizamos la fórmula de estandarización:
   $$
   Z = \frac{X - \mu}{\sigma}
   $$
   En este caso, queremos calcular $P(X < 85)$. Primero transformamos $X = 85$ a la variable estandarizada $Z$:
   $$
   Z = \frac{85 - 70}{10} = 1.5
   $$

2. **Calcular la probabilidad usando la distribución normal estándar**:
   Ahora que tenemos $Z = 1.5$, podemos utilizar las tablas de la distribución normal estándar o una herramienta computacional para encontrar $P(Z < 1.5)  \approx 0.9332$. Por lo tanto, la probabilidad de que un estudiante obtenga una puntuación menor que 85 es aproximadamente 0.9332.

   

En realidad, las tablas de distribución están perdiendo relevancia ante la facilidad obtenida por las herramientas computacionales, las cuales posibilitan la obtención de resultados con una mayor precisión y rapidez.

En Python podemos realizar el cálculo sin necesidad de estandarizar:

```python
# Import librerias
from scipy.stats import norm

prob_X_less_85 = norm.cdf(85,70,10)
print(f'P(X<85)= {prob_X_less_85}')
#Notemos como al estandarizar se obtiene el mismo resultado
prob_X_less_85 = norm.cdf(1.5,0,1)
print(f'P(X<85)= {prob_X_less_85}')
```

En la fabricación de cereales, se contempla el peso de las cajas como una variable aleatoria con distribución normal. Una marca reconocida presenta cajas cuyo peso promedio es de 1000 gramos y una desviación estándar de 20 gramos. Se selecciona una caja al azar con el propósito de realizar un análisis detallado.

Determinar la probabilidad de que pese menos de 1025 gramos.

Hallar la probabilidad de que pese 989 gramos o más.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros de la distribución normal
media = 1000
desviacion_estandar = 20

# Valor para el cual queremos calcular la probabilidad (X < 1025)
valor_limite1 = 1025
valor_limite2 = 989

# Calcula la probabilidad utilizando la función de distribución acumulativa (CDF)
probabilidad1 = norm.cdf(valor_limite1, loc=media, scale=desviacion_estandar)
probabilidad2 = (1-norm.cdf(valor_limite2, loc=media, scale=desviacion_estandar))

print(f"La probabilidad de que pese  menos de {valor_limite1} onzas  es: {probabilidad1:.1%}")
print(f"La probabilidad de que pese  mas de {valor_limite2} onzas  es: {probabilidad2:.1%}")
```

```python
# Generar datos para el gráfico
x = np.linspace(media - 3 * desviacion_estandar, media + 3 * desviacion_estandar, 1000)
y = norm.pdf(x, loc=media, scale=desviacion_estandar)

# Grafica la distribución normal
plt.plot(x, y, label='Distribución Normal')

# Sombréa la región P(X < 1025)
x_fill = np.linspace(media - 3 * desviacion_estandar, valor_limite1, 1000)
y_fill = norm.pdf(x_fill, loc=media, scale=desviacion_estandar)
plt.fill_between(x_fill, y_fill, color='lightblue', alpha=0.5, label='P(X < 1025)')

# Resalta la media
plt.axvline(media, color='red', linestyle='--', label='Media')

# Configura el gráfico
plt.title('Distribución Normal y P(X < 1025)')
plt.xlabel('X')
plt.ylabel('Densidad de Probabilidad')
plt.legend()

# Muestra el porcentaje asociado
plt.text(1002, 0.015, f'P(X < 1025) = {probabilidad1:.2%}', fontsize=10, color='darkblue')

# Muestra el gráfico
plt.show()
```

## Distribución Chi-cuadrado ($\chi^2$) como la suma de variables normalmente distribuidas

En estadística, la distribución chi-cuadrado con _k_ grados de libertad es una distribución de una suma de cuadrados de _k_ variables independientes normalmente estandarizadas.

Se usa mucho en inferencia, para pruebas de hipótesis e intervalos de confianza.

La PDF de la distribución chi-cuadrado es:

$$ f(x;\,k)={\begin{cases}{\dfrac {x^{{\frac {k}{2}}-1}e^{-{\frac {x}{2}}}}{2^{\frac {k}{2}}\Gamma \left({\frac {k}{2}}\right)}},&x>0;\\0,&{\text{otros}}.\end{cases}} $$

donde $\Gamma({k/2})$ es la función gamma, que tiene valores cerrados para los enteros $k$.

### Ejemplo 3: Caminata aleatoria en una dimensión (Distribución normal)

Considere el problema de la caminata aleatoria de una dimensión, el cual consite en determinar un punto fijo de inicio y se debe dar un paso a la derecha o a la izquierda con la misma probabilidad. Luego de varios pasos fijos, se determina la posición final. Esta caminata se repite varias veces para establecer un histograma de distancias, el cual seguirá una distribución normal.

```python
# Python code for 1-D random walk.
import random
import numpy as np
import matplotlib.pyplot as plt

# Probability to move up or down
prob = [0.5, 0.5] 

# statically defining the starting position
start = 2
positions = [start]
n_steps = 100
# creating the random points
rr = np.random.random(n_steps)
downp = rr < prob[0]
upp = rr > prob[1]


for idownp, iupp in zip(downp, upp):
	down = idownp and positions[-1] > 1
	up = iupp and positions[-1] < 4
	positions.append(positions[-1] - down + up)

# plotting down the graph of the random walk in 1D
plt.plot(positions)
plt.title(f'Random Walk starting at {start} with {n_steps} steps' )
plt.show()
```

```python
#  Caminata aleatoria 1D

import random
import numpy as np
import matplotlib.pyplot as plt

def random_walk(steps, step_size):
    walk = 0
    for i in range(steps):
        walk += step_size*random.choice([-1,1])
    return walk

num_walks = 8000
num_steps = 100 
step_size = 0.5

#  Realización de varias caminatas y almacenamiento de la distancia total.

last_pos = np.zeros(num_walks)
for i in range(num_walks):      
    walk = random_walk(num_steps,step_size)
    last_pos[i] = walk


mean_dist = np.average(last_pos);


print(' Para %d pasos, la distancia promedio al centro es = %5.1f' %(num_steps, mean_dist))
# Gráfica
plt.title('Histograma de posiciones finales #num_walks = %i, #steps = %i, mean_dist = %6.1f' % (num_walks,num_steps, mean_dist))
plt.hist(last_pos, bins=50, color='red') 
plt.xlabel('Número de pasos')
plt.ylabel('Frecuencia')
plt.show() 
```

## Ejemplo ( Distribución normal inversa)

Se utilizan medidores para rechazar todos los componentes en los que cierta dimensión
no esté dentro de la especificación 1.50 ± d. Se sabe que esta medida se distribuye normalmente con una media de 1.50 y una desviación estándar de 0.2. Determine el valor d tal que las especificaciones “cubran” 95% de las mediciones.

(Tomado de Walpole R., et al. Probabilidad y estadística para ingeniería y ciencias. Página 184 ( Ejemplo 6.10)

### Solución.

Sea $X$ la variable aleatoria que determina las medidas de un componente aleatorio. Entonces $X$ tiene una distribución normal con media 1.50 y desviación estándar de 0.2. Debemos hallar el valor d tal que 

$$P( -d<X-1.50<d)=0.95 $$

Equivalentemente: 
$$P( -d+1.5<X<d+1.5)=0.95 $$

Podemos estándarizar, para obtener: 
$$P\left( \frac{-d}{0.2}<Z=\frac{X-1.5}{0.2}< \frac{d}{0.2}\right)=0.95 $$





Para hallar este valor d podemos utilizar la función `stats.norm.ppf()`

**Nota:** 






La función `stats.norm.ppf` proviene del módulo `scipy.stats` de Python, y se utiliza para calcular el **cuantil** o el **valor inverso de la función de distribución acumulativa (CDF)** de una distribución normal estándar.

 
**`ppf` (Percent Point Function):**  
   Es la función inversa de la CDF.  
   Dado un porcentaje acumulado $ p $, `ppf` devuelve el valor $z $ en el eje de la distribución normal estándar tal que la probabilidad acumulada desde la cola izquierda hasta $ z $ es igual a $ p $.

   En otras palabras:
  $$
   P(Z \leq z) = p
   $$



```python
import scipy.stats as stats

# Datos del problema
media = 1.50
desviacion_estandar = 0.2
porcentaje = 0.95

# Cálculo de d
z = stats.norm.ppf(1-0.025)  # Z para un intervalo simétrico de 95%
d = z * desviacion_estandar

print(f"El valor de d que cubre el 95% de las mediciones es: {d:.4f}")
```
