```python
# Importamos las librerias necesarias
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
 

#Cargamos la base de datos previamente descargada en la carpeta de trabajo.
df = pd.read_csv("obesity_levels.csv")

```

```python
df_male= df[df['Gender']=='Male']

height = df_male['Height']

height.hist(bins= 12)
plt.show()
```

```python
# Seleccionamos el tamaÃ±o para cada muestra
tamaÃ±o_muestra = 20

# Creamos las 1000 muestras aleatorias, cada una de tamaÃ±o 
#tamaÃ±o_muestra, de la columna Weight para las mujeres
muestras = [height.sample(tamaÃ±o_muestra) for _ in range(100)]

# Calcula el promedio de cada muestra
promedios_muestras = [muestra.mean() for muestra in muestras]
# Calculemos el promedio del promedio de las muestras
promedios = np.mean(promedios_muestras)

# Crea un histograma del promedio de las muestras
plt.hist(promedios_muestras, bins=20, edgecolor='black')    
plt.title(f'Histograma del Promedio de estatura promedio del peso de las mujeres (n={tamaÃ±o_muestra})')
plt.xlabel('Promedio de Muestra' )
plt.ylabel('Frecuencia')
plt.axvline(x=promedios,color='r') # for giving mean line
plt.show()

```
