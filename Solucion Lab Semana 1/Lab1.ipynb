```python
# importamos paquetes

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statistics
```

```python
# Cargamos los datos

data = pd.read_csv('realistic_ocean_climate_dataset.csv')
```

```python
data.head()
```

```python
data_ph = data['pH Level']
```

```python
# mean, media, mode, mediana, Q1, Q3 and standard desviation

print(f'The value of the mean is {np.mean(data_ph):.3f}')
print(f'The value of the median is {np.median(data_ph):.3f}')
print(f'The value of the mode is {statistics.mode(data_ph):.3f}')
print(f'The value of the standard deviation is {np.std(data_ph):.3f}')

print(f'The value of the quartil 1 Q1 is {np.quantile(data_ph,0.25)}')
print(f'The value of the quartil 3 Q3 is {np.quantile(data_ph,0.75)}')
```

```python
# Diagrama de cajas

plt.boxplot(data_ph,vert = False)
plt.title('Ph Level')
plt.show()
```

# Respuestas a las preguntas

Observamos que la media y la mediana son iguales(8.056), lo que muestra una simetria en la dispersión de los datos. La desviación estándar de 0.06 muestran que los datos tienen una dispersión baja alrededor de la media.

El cuartil 1 nos muestra que el 25% de los datos es menor a 8.011 y el cuartil 3 muestra que el 75% de las mediciones de Ph es menor a 8.08525.

Por útimo, el diagrama de cajar muestra que hay dos valores atipicos en las mediciones de Ph level. Esto corresponde a dos mediciones con un bajo Ph level(niveles de Ph bajo significa que el nivel de acidez del agua es alto). Adiciónalmente, observemos que la caja con respecto a la mediana, es mas ancha hacia la izquierda, lo que muestra que hay una tendencia de valores menores a la mediana.
