```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('student_habits_performance.csv')
```

# Atividade 1

```python
var_principal = 'social_media_hours'

df_filtrado = df[df['gender'].isin(['Male', 'Female'])].copy()

df_mulheres = df_filtrado[df_filtrado['gender'] == 'Female'][var_principal]
df_homens = df_filtrado[df_filtrado['gender'] == 'Male'][var_principal]

NUM_BINS = 8
min_global = df_filtrado[var_principal].min()
max_global = df_filtrado[var_principal].max()

bin_edges = np.linspace(min_global, max_global, NUM_BINS + 1)

print("--- Análise da Variável 'social_media_hours' ---")
print(f"Valor Mínimo Global: {min_global:.2f} horas")
print(f"Valor Máximo Global: {max_global:.2f} horas\n")
```

```python
counts_mulheres, _ = np.histogram(df_mulheres, bins=bin_edges)

tabela_mulheres = pd.DataFrame({
    'Limite Inferior': bin_edges[:-1].round(2),
    'Limite Superior': bin_edges[1:].round(2),
    'Frequência Absoluta': counts_mulheres
})
tabela_mulheres['Ponto Médio'] = (tabela_mulheres['Limite Inferior'] + tabela_mulheres['Limite Superior']) / 2

print("--- Tabela de Frequências - Mulheres ---")
print(tabela_mulheres[['Limite Inferior', 'Limite Superior', 'Ponto Médio', 'Frequência Absoluta']])
print("\n")
```

```python
counts_homens, _ = np.histogram(df_homens, bins=bin_edges)

tabela_homens = pd.DataFrame({
    'Limite Inferior': bin_edges[:-1].round(2),
    'Limite Superior': bin_edges[1:].round(2),
    'Frequência Absoluta': counts_homens
})
tabela_homens['Ponto Médio'] = (tabela_homens['Limite Inferior'] + tabela_homens['Limite Superior']) / 2

print("--- Tabela de Frequências - Homens ---")
print(tabela_homens[['Limite Inferior', 'Limite Superior', 'Ponto Médio', 'Frequência Absoluta']])
print("\n")
```

# Atividade 2

```python
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.hist(df_homens, bins=bin_edges, color='#1f77b4', edgecolor='black', alpha=0.8) # Azul
plt.title('Horas de Mídia Social - Homens')
plt.xlabel('Horas por Dia')
plt.ylabel('Frequência Absoluta')
plt.grid(axis='y', alpha=0.5)

plt.subplot(1, 2, 2)
plt.hist(df_mulheres, bins=bin_edges, color='#ff7f0e', edgecolor='black', alpha=0.8) # Laranja
plt.title('Horas de Mídia Social - Mulheres')
plt.xlabel('Horas por Dia')
plt.ylabel('Frequência Absoluta')
plt.grid(axis='y', alpha=0.5)

plt.tight_layout()
plt.show()
```

# Atividade 3

```python
print("--- Descrição e Interpretação dos Gráficos ---")

print("\n1. **Análise da Distribuição (Forma):**")
print("""   - Ambos os grupos apresentam uma distribuição com **assimetria positiva** 
(ou cauda à direita), o que é comum para variáveis de tempo de uso. A maioria dos 
estudantes passa menos tempo nas mídias sociais, mas existe uma minoria que reporta 
um uso muito elevado.""")

print("\n2. **Semelhanças (Tendência Central):**")
print("""   - A **moda** (o pico de frequência) para ambos os grupos se encontra na 
mesma faixa, aproximadamente entre 2.0 e 3.0 horas por dia. Isso indica que o padrão 
de uso mais comum é semelhante independentemente do gênero.""")

print("\n3. **Diferenças (Frequências nos Extremos):**")
print("""   - Observando as tabelas e os gráficos, as **mulheres** tendem a ter 
frequências absolutas mais elevadas nos intervalos de uso mais alto 
(acima de 4.0 ou 5.0 horas). Isso sugere que a proporção de estudantes do sexo 
feminino que são usuárias intensivas de mídia social é maior na amostra, em 
comparação com os homens.""")
```

# Atividade 4

```python
var_relacionada = 'exam_score'

correlation_coefficient = df[var_principal].corr(df[var_relacionada])

plt.figure(figsize=(8, 6))

plt.scatter(df[var_principal], df[var_relacionada], alpha=0.5, color='darkred', marker='o')
plt.title(f'Diagrama de Dispersão: Horas de Mídia Social vs. Nota de Exame')
plt.xlabel('Horas de Mídia Social por Dia')
plt.ylabel('Nota de Exame')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

print("--- Diagrama de Dispersão e Interpretação ---")

print(f"\n Variável Quantitativa Selecionada: **{var_relacionada}**")

print(f"\n Coeficiente de Correlação (ρ): **{correlation_coefficient:.3f}**")

print("\n1. **Análise do Diagrama de Dispersão (Forma):**")
print("""   - O gráfico de dispersão mostra uma **nuvem de pontos amplamente dispersa** 
que se estende por todo o intervalo de notas de exame para quase todos os níveis de uso 
de mídia social.""")

print("\n   - Não há um padrão linear óbvio (reta) nem positivo nem negativo visível no gráfico.")

print("\n2. **Interpretação da Relação (Correlação):**")
print(f"""\n   - O Coeficiente de Correlação é de **{correlation_coefficient:.3f}**, 
um valor muito próximo de zero (próximo a 0.5 é moderado, próximo a 1.0 é forte).""")

print("""\n   - Baseado no gráfico desorganizado e no baixo valor de ρ, existe 
uma **correlação linear muito fraca ou inexistente** entre o tempo gasto em mídias 
sociais e o desempenho acadêmico (Nota de Exame) nesta amostra.""")

print("""\n   - Embora seja intuitivo que mais mídias sociais levem a notas mais 
baixas (relação negativa), o gráfico e o coeficiente sugerem que, para esta amostra, 
a variação nas notas de exame é explicada por fatores diferentes ou mais fortes do 
que apenas as horas de mídia social (como horas de estudo ou saúde mental).""")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('Final_data.csv')
df.head()
```

```python
df = pd.read_csv('Final_data.csv')
```

```python
freq_workout = df['Workout_Type'].value_counts()

plt.figure(figsize=(10, 5))
freq_workout.plot(kind='bar', color='green', edgecolor='black', alpha=0.7)
plt.title('Gráfico de Barras: Tipos de Treino (Workout_Type)')
plt.xlabel('Tipo de Treino')
plt.ylabel('Frequência Absoluta')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 8))
percent_workout = freq_workout / freq_workout.sum() * 100

plt.pie(
    percent_workout, 
    labels=freq_workout.index, 
    autopct='%1.1f%%', 
    startangle=90, 
    colors=plt.cm.Paired.colors, 
    wedgeprops={'edgecolor': 'black'}
)
plt.title('Gráfico de Pizza: Distribuição de Tipos de Treino')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
var_proteinas = 'Proteins'

min_proteins = df[var_proteinas].min()
max_proteins = df[var_proteinas].max()

n = len(df)
k = int(np.ceil(1 + 3.322 * np.log10(n)))
NUM_BINS_P = k

bin_edges_proteins = np.linspace(min_proteins, max_proteins, NUM_BINS_P + 1)

counts_proteins, _ = np.histogram(df[var_proteinas], bins=bin_edges_proteins)

tabela_proteinas = pd.DataFrame({
    'Limite Inferior (g)': bin_edges_proteins[:-1].round(2),
    'Limite Superior (g)': bin_edges_proteins[1:].round(2),
    'Frequência Absoluta': counts_proteins
})

tabela_proteinas['Ponto Médio (g)'] = (tabela_proteinas['Limite Inferior (g)'] + tabela_proteinas['Limite Superior (g)']) / 2

tabela_proteinas['Frequência Relativa (%)'] = (tabela_proteinas['Frequência Absoluta'] / n * 100).round(2)
tabela_proteinas['Frequência Acumulada'] = tabela_proteinas['Frequência Absoluta'].cumsum()


tabela_proteinas
```

```python
plt.figure(figsize=(9, 5))
plt.hist(df[var_proteinas], bins=bin_edges_proteins, color='#8c564b', edgecolor='black', alpha=0.8)
plt.title(f'Histograma da Quantidade de Proteínas Consumidas')
plt.xlabel('Proteínas (g)')
plt.ylabel('Frequência Absoluta')
plt.grid(axis='y', alpha=0.5)
plt.show()
```

```python
print("**1. Interpretação de 'Workout_Type' (Gráficos de Barras e Pizza):**")
print(f"\n   - Os gráficos de Barras e Pizza mostram a distribuição dos tipos de treino.")
print(f"""\n   - O tipo de treino **'{freq_workout.index[0]}'** é claramente o mais frequente, 
      representando **{percent_workout.iloc[0]:.1f}%** da amostra.""")
print(f"""\n   - O tipo de treino **'{freq_workout.index[-1]}'** é o menos comum. Esta variável 
      é **qualitativa nominal**, e a interpretação se baseia na simples comparação da frequência 
      de cada categoria.""")

print(f"\n**2. Interpretação de 'Proteins' (Tabela e Histograma):**")
print(f"""\n   - **Forma:** O histograma das Proteínas mostra uma distribuição que tende a ser 
      **simétrica ou levemente assimétrica**.""")
print(f"""\n   - **Tendência Central:** A maior concentração de consumo de Proteínas (Moda) se 
      encontra na faixa de **{tabela_proteinas['Limite Inferior (g)'].iloc[0]:.2f}g a {tabela_proteinas['Limite Superior (g)'].iloc[0]:.2f}g** (o primeiro intervalo da tabela).""")
print("""\n   - **Dispersão:** A distribuição se concentra fortemente nas faixas centrais, indicando 
      que a maioria dos indivíduos da amostra consome uma quantidade de Proteínas dentro de um limite 
      relativamente estreito (entre 60g e 130g, por exemplo), com poucas observações nos extremos 
      (consumo muito baixo ou muito alto).""")
```
