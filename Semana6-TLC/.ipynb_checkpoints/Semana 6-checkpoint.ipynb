### Ejemplo
Generemos muestras aleatorias de números entre 0 y 10 y representemos gráficamente su distribución para diferentes tamaños de muestra: n=1, n=10, n=50 y n=100.


```python
import numpy as np
import matplotlib.pyplot as plt

# Establecer tamaños de muestra
sample_sizes = [1, 10, 50, 100]

# Crear una figura y ejes para los gráficos
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.flatten()

# Generar y graficar muestras aleatorias
for i, n in enumerate(sample_sizes):
    samples = np.random.randint(0, 11, size=n)
    axes[i].hist(samples, bins=range(12), edgecolor='black', alpha=0.7)
    axes[i].set_title(f'Sample Size: {n}')
    axes[i].set_xlabel('Value')
    axes[i].set_ylabel('Frequency')

# Ajustar el layout y mostrar el gráfico
plt.tight_layout()
plt.show()
```

Luego calcularemos la media muestral para cada tamaño de la muestra y observaremos el comportamiento de su distribución.


```python
import numpy
import matplotlib.pyplot as plt

#Seleccionamos varios tamaño para la muestra
num = [1, 10, 50, 100] 
# list of sample means
means = [] 

# A continuación se generan 1, 10, 200, 500 números aleatorios entre 0 y 10
# Calculamos su media.
    
# Usamos la función seed de manera que cada vez que se ejecute el bucle
# obtengamos el mismo resultado.
for j in num:
	numpy.random.seed(20)
	x = [numpy.mean(
		numpy.random.randint(
			0, 10, j)) for _i in range(1000)]
	means.append(x)
k = 0

# Dibujamos las cuatro figuras.
fig, ax = plt.subplots(2, 2, figsize =(8, 8))
for i in range(0, 2):
	for j in range(0, 2):
		# Histogram for each x stored in means
		ax[i, j].hist(means[k], 10, density = True)
		ax[i, j].set_title(label = num[k])
		k = k + 1
plt.show()
```

## Teorema del Límite Central
Para una muestra aleatoria de tamaño 𝒏 de una población con media 𝝁 y desviación estándar 𝛔,  la distribución  de la media muestral $\bar{𝒙}$  tiene una media aproximadamente  igual a la media de la población 𝝁 y una desviación estándar igual a la desviación estándar de la población dividida por la raíz cuadrada del tamaño de la muestra, es decir, 

$$\frac{\sigma}{\sqrt{n}}$$

### Interpretación del TLC
Cuando se eligen muestras de un tamaño específico de una población, la distribución de las medias de estas muestras tiende a seguir una distribución normal. Esta aproximación se vuelve más precisa a medida que aumenta el tamaño de las muestras. Este comportamiento se observa en las siguientes tres gráficas.



```python
#Código para generar las gráficas

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Definir la población (puedes cambiar la distribución según tus necesidades)
poblacion = np.random.normal(loc=0, scale=1, size=100000)

# Función para generar muestras y calcular la media
def generar_muestras(tamano_muestra):
    medias_muestrales = [np.mean(np.random.choice(poblacion, size=tamano_muestra)) for _ in range(1000)]
    return medias_muestrales

# Tamaños de muestra
tamanos_muestra = [10, 50, 100]

# Crear subgráficos
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Generar gráficos para diferentes tamaños de muestra
for i, n in enumerate(tamanos_muestra):
    medias_muestrales = generar_muestras(n)
    axs[i].hist(medias_muestrales, bins=30, density=True, alpha=0.7, color='skyblue', edgecolor='red')
    
    # Añadir la curva de la distribución gaussiana
    xmin, xmax = axs[i].get_xlim()
    x = np.linspace(xmin, xmax, 100)
    p = norm.pdf(x, np.mean(medias_muestrales), np.std(medias_muestrales))
    axs[i].plot(x, p, 'k', linewidth=2)
    
    axs[i].set_title(f'Tamaño de muestra n = {n}')
    axs[i].set_xlabel('Media muestral')
    axs[i].set_ylabel('Frecuencia relativa')

plt.tight_layout()
plt.show()
```

Suposiciones detrás del Teorema del Límite Central

Para usar el modelo normal, debemos cumplir con algunas suposiciones y condiciones. El Teorema del Límite Central asume lo siguiente:

- **Condición de Aleatoriedad**: Los datos deben ser muestreados aleatoriamente.

- **Suposición de Independencia**: Los valores de la muestra deben ser independientes entre sí. Esto significa que la ocurrencia de un evento no tiene influencia sobre el siguiente evento. Por lo general, si sabemos que las personas o ítems fueron seleccionados aleatoriamente, podemos asumir que se cumple la suposición de independencia.

- **Condición del 10%**: Cuando la muestra se extrae sin reemplazo (generalmente es el caso), el tamaño de la muestra, n, no debe ser más del 10% de la población.

- **Suposición de Tamaño de la Muestra**: Ahora, ¿cómo determinaremos qué tan grande debe ser este tamaño? Bueno, depende de la población. Cuando la población está sesgada o es asimétrica, el tamaño de la muestra debe ser grande. Si la población es simétrica, entonces podemos tomar muestras pequeñas también.

En general, un tamaño de muestra de 30 se considera suficiente cuando la población es simétrica.

### Ejemplo:

Validaremos el teorema del límite central utilizando una base de datos real.

Consideremos la base de datos Used Cars Dataset (un conjunto de datos que contiene anuncios de automóviles con numerosas características categóricas y numéricas). En particular,  nos enfocaremos en la variable aleatoria usd_price.  A continuación se detallan los pasos a seguir: 

**Paso 1:** Escojamos inicialmente como n=5 el tamaño de la muestra aleatoria de la variable price_usd. 

**Paso 2:** Realizaremos 1000 muestras aleatorias con un tamaño de muestra de 5 de la columna price_usd y calcularemos el promedio de cada muestra.

**Paso 3:** : Usaremos la biblioteca Matplotlib para generar un histograma que ilustre la distribución de los 1000 promedios de muestras de tamaño 5, estableciendo el parámetro bins en 30. 

Verificaremos que la distribución de las medias de estas muestras tiende a seguir una distribución normal. Esta aproximación se volverá más precisa a medida que aumentemos el tamaño de las muestras, por lo que repetiremos el cálculo tomando muestras de tamaño 10 y 30. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
 

# Cargamos la base de datos cars.csv, previamente guardada en la carpeta de trabajo:
#https://www.kaggle.com/datasets/austinreese/craigslist-carstrucks-data
archivo_csv = 'cars.csv'

# Lee el archivo CSV
data_frame = pd.read_csv(archivo_csv)

# Columna que deseamos leer
nombre_columna = 'price_usd'

# Accedemos a la columna específica
columna_seleccionada = data_frame[nombre_columna]
media_price =np.mean(columna_seleccionada)
# Reemplaza 'tamaño_muestra' con el tamaño deseado para cada muestra
tamaño_muestra = 20

# Creamos las 1000 muestras aleatorias, cada una de tamaño tamaño_muestra, de la columna price_usd
muestras = [columna_seleccionada.sample(tamaño_muestra) for _ in range(1000)]

# Calcula el promedio de cada muestra
promedios = [muestra.mean() for muestra in muestras]

# Crea un histograma del promedio de las muestras
plt.hist(promedios, bins=30, edgecolor='black')  # Puedes ajustar el número de bins con otros valores
plt.title(f'Histograma del Promedio de Muestras price_usd (n={tamaño_muestra})')
plt.xlabel('Promedio de Muestra' )
plt.ylabel('Frecuencia')
plt.axvline(x=media_price,color='r') # for giving mean line
plt.show()


```

### Ejemplo:

Las latas de comida se envasa en vastas líneas de ensamblaje. Un tamaño particular de lata de sopa debe tener 19oz, pero hay variabilidad de una lata a otra – es una variable aleatoria. Supongamos que la distribución del contenido es aproximadamente normal con una media de 19.1 oz y una desviación estándar de 0.05.

a) Encuentra la probabilidad de que una lata seleccionada al azar contenga menos de 19oz. 

b) Los almacenes mayoristas venden estas latas en paquetes de cuatro. Encuentra la probabilidad de que el promedio de un paquete seleccionado al azar de cuatro latas sea inferior a 19oz.  

c) ¿Entre qué par de valores encontrarías el contenido promedio del 95% central de los paquetes de cuatro latas?

d) ¿Por qué podríamos usar la distribución normal para encontrar las probabilidades en este problema?

### Solución:

a) En este caso una lata de comida tiene una distribución normal con media $\mu=19.1$ y desviación estándar de $0.05$. Por tanto la probabilidad de que una lata seleccionada al azar contenga menos de 19 oz es

$$P(X<19)= 0.022$$

Es decir, la probabilidad de que una lata seleccionada al azar contenga menos de 19 oz es $2,3\%$

b) En este caso, podemos aplicar el teorema del límite central que afirma que si $\bar{X}_4$ denota el contenido promedio de una muestra aleatoria de 4 latas entonces $\bar{X}_4$ tiene distribución normal con media $\mu = 19.1$ y desviación estándar 
$$\sigma = \frac{0.05}{\sqrt{4}}$$.

Por tanto $$P(\bar{X}_4<19)=0.000032$$

La probabilidad de que el promedio de un paquete seleccionado al azar de cuatro latas sea inferior a 19oz es $0.0032\%$ 

c) Podemos usar la regla empirica de la desviación estándar que afirma que el 95% de los datos se encuentran alrededor de la media una ditancia de dos desviaciones estándar:

<img src="normaldistribution.png">


Por tanto, dado que la desviación estándar de $\bar{X}_4$ es  $\sigma =0.025 $ el par de valores donde se encontrará el contenido promedio del 95% central de los paquetes de cuatro latas es

$$19.1- 2*0.025 = 19.05$$
$$19.1+ 2*0.025 = 19.15$$

d) Podemos usar la distribución normal dado que el teorema del límite central permite concluir que $\bar{X}_4$ tiene una distribución normal dado que $X$ tiene una distribución normal a pesar de que el tamaño de la muestra es pequeño.

```python
# Cálculos necesarios para resolver el problema anterior.

#importando el módulo scipy.stats, que proporciona una variedad de funciones y distribuciones estadísticas.
import scipy.stats as stats

mean = 19.1
dev = 0.05
# a) 
x= 19
prob_less_than19= stats.norm.cdf( x, mean, dev)
print(f'P(X<19)={prob_less_than19:.3f}')

#b) 
mean = 19.1
dev = 0.05/(4)**(0.5)
x= 19

prob_less_than19= stats.norm.cdf( x, loc= mean, scale = dev)
print(f'P(X_4 <19)={prob_less_than19:.6f}')

# c) 
print(dev)
```

### Ejemplo: 

Supongamos que **X** representa el número de pacientes con ataques cardíacos que llegan a una clínica por semana. La distribución de **X** tiene una media de 3.5 pacientes y una desviación estándar de 1.7.

a) ¿Puede la variable aleatoria **X** tener una distribución binomial?
   - No, la variable aleatoria $X$ no tiene las caracteristicas de una distribución binomial. Por ejemplo, no hay un número de ensayos y tampoco una probabilidad de éxito.

b) ¿Puede la variable aleatoria **X** tener una distribución normal?
   - No, la distribución de X debe estar sesgada hacia la derecha.

c) Supongamos que estamos interesados en $\bar{X}$, que es el número promedio de ataques cardíacos por semana en un año. ¿Cuál sería el valor de **n**?
   - **n = 52**. Este valor corresponde al número de semanas en un año.

d) Encuentra la distribución de la media muestral.
   - Dado que **n = 52** es lo suficientemente grande, $\bar{X}$ sigue una distribución normal con media de 3.5 y desviación estándar de:
     $$\frac{1.7}{\sqrt{52}} = 0.236$$

e) Encuentra la probabilidad de que, en un año, haya un promedio de menos de 3 ataques cardíacos por semana en la clínica.
   - Respuesta:
   Dado que $\bar{X}$ sigue aproximandamente una distribución normal, con media de 3.5 y desviación estándar de 0.236, obtenemos que :
     $$P(\bar{X} < 3) = P\left(Z < \frac{3 - 3.5}{0.236}\right) = P(Z < -2.12) = 0.0170$$
     Aproximadamente el 2% de los años tendrán un promedio de menos de 3 ataques cardíacos.

### Ejemplo: 
Hay 40 estudiantes en una clase de estadística elemental. Basado en años de experiencia, el instructor sabe que el tiempo requerido para calificar un primer examen seleccionado al azar es una variable aleatoria con un valor esperado de 6 min y una desviación estándar de 6 min. Si los tiempos de calificación son independientes y el instructor comienza a calificar a las 6:50 p.m., y califica en forma continua, 

a) ¿cuál es la probabilidad (aproximada) de que termine de calificar antes de que se inicie el programa de noticias de las 11:00 p.m.?  

b)Si el reporte de deportes se inicia a la 11:10, ¿cuál es la probabilidad de que se pierda una parte del reporte si espera
hasta que termine de calificar para prender la TV?
(Tomado de Devore, Probabilidad p. 229).



```python
# Cálculos necesarios para resolver el problema anterior.

#importando el módulo scipy.stats, que proporciona una variedad de funciones y distribuciones estadísticas.
import scipy.stats as stats

mean = 6
dev = 6/(40**0.5)
#a) 
#Notemos que desde las 6:50 hasta las 11:00 tenemos 250 minutos. Por estudiante se tiene 250/40=6.25

x= 6.25
prob_less_than= stats.norm.cdf( x, mean, dev)
print(f'P(X<6.25)={prob_less_than:.4}')

#b)
#Notemos que desde las 6:50 hasta las 11:10 tenemos 260 minutos. Por estudiante se tiene 260/40=6.5
x = 6.5
prob_greather_than= 1- stats.norm.cdf( x, mean, dev)
print(f'P(X>6.5)={prob_greather_than:.4}')
```
