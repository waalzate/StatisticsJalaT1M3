### Ejemplo
Generemos muestras aleatorias de n칰meros entre 0 y 10 y representemos gr치ficamente su distribuci칩n para diferentes tama침os de muestra: n=1, n=10, n=50 y n=100.


```python
import numpy as np
import matplotlib.pyplot as plt

# Establecer tama침os de muestra
sample_sizes = [1, 10, 50, 100]

# Crear una figura y ejes para los gr치ficos
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.flatten()

# Generar y graficar muestras aleatorias
for i, n in enumerate(sample_sizes):
    samples = np.random.randint(0, 11, size=n)
    axes[i].hist(samples, bins=range(12), edgecolor='black', alpha=0.7)
    axes[i].set_title(f'Sample Size: {n}')
    axes[i].set_xlabel('Value')
    axes[i].set_ylabel('Frequency')

# Ajustar el layout y mostrar el gr치fico
plt.tight_layout()
plt.show()
```

Luego calcularemos la media muestral para cada tama침o de la muestra y observaremos el comportamiento de su distribuci칩n.


```python
import numpy
import matplotlib.pyplot as plt

#Seleccionamos varios tama침o para la muestra
num = [1, 10, 50, 100] 
# list of sample means
means = [] 

# A continuaci칩n se generan 1, 10, 200, 500 n칰meros aleatorios entre 0 y 10
# Calculamos su media.
    
# Usamos la funci칩n seed de manera que cada vez que se ejecute el bucle
# obtengamos el mismo resultado.
for j in num:
	numpy.random.seed(20)
	x = [numpy.mean(
		numpy.random.randint(
			0, 10, j)) for _i in range(1000)]
	means.append(x)
k = 0

# Dibujamos las cuatro figuras.
fig, ax = plt.subplots(2, 2, figsize =(8, 8))
for i in range(0, 2):
	for j in range(0, 2):
		# Histogram for each x stored in means
		ax[i, j].hist(means[k], 10, density = True)
		ax[i, j].set_title(label = num[k])
		k = k + 1
plt.show()
```

## Teorema del L칤mite Central
Para una muestra aleatoria de tama침o 洧눒 de una poblaci칩n con media 洧례 y desviaci칩n est치ndar 洧띖,  la distribuci칩n  de la media muestral $\bar{洧눛}$  tiene una media aproximadamente  igual a la media de la poblaci칩n 洧례 y una desviaci칩n est치ndar igual a la desviaci칩n est치ndar de la poblaci칩n dividida por la ra칤z cuadrada del tama침o de la muestra, es decir, 

$$\frac{\sigma}{\sqrt{n}}$$

### Interpretaci칩n del TLC
Cuando se eligen muestras de un tama침o espec칤fico de una poblaci칩n, la distribuci칩n de las medias de estas muestras tiende a seguir una distribuci칩n normal. Esta aproximaci칩n se vuelve m치s precisa a medida que aumenta el tama침o de las muestras. Este comportamiento se observa en las siguientes tres gr치ficas.



```python
#C칩digo para generar las gr치ficas

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Definir la poblaci칩n (puedes cambiar la distribuci칩n seg칰n tus necesidades)
poblacion = np.random.normal(loc=0, scale=1, size=100000)

# Funci칩n para generar muestras y calcular la media
def generar_muestras(tamano_muestra):
    medias_muestrales = [np.mean(np.random.choice(poblacion, size=tamano_muestra)) for _ in range(1000)]
    return medias_muestrales

# Tama침os de muestra
tamanos_muestra = [10, 50, 100]

# Crear subgr치ficos
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Generar gr치ficos para diferentes tama침os de muestra
for i, n in enumerate(tamanos_muestra):
    medias_muestrales = generar_muestras(n)
    axs[i].hist(medias_muestrales, bins=30, density=True, alpha=0.7, color='skyblue', edgecolor='red')
    
    # A침adir la curva de la distribuci칩n gaussiana
    xmin, xmax = axs[i].get_xlim()
    x = np.linspace(xmin, xmax, 100)
    p = norm.pdf(x, np.mean(medias_muestrales), np.std(medias_muestrales))
    axs[i].plot(x, p, 'k', linewidth=2)
    
    axs[i].set_title(f'Tama침o de muestra n = {n}')
    axs[i].set_xlabel('Media muestral')
    axs[i].set_ylabel('Frecuencia relativa')

plt.tight_layout()
plt.show()
```

Suposiciones detr치s del Teorema del L칤mite Central

Para usar el modelo normal, debemos cumplir con algunas suposiciones y condiciones. El Teorema del L칤mite Central asume lo siguiente:

- **Condici칩n de Aleatoriedad**: Los datos deben ser muestreados aleatoriamente.

- **Suposici칩n de Independencia**: Los valores de la muestra deben ser independientes entre s칤. Esto significa que la ocurrencia de un evento no tiene influencia sobre el siguiente evento. Por lo general, si sabemos que las personas o 칤tems fueron seleccionados aleatoriamente, podemos asumir que se cumple la suposici칩n de independencia.

- **Condici칩n del 10%**: Cuando la muestra se extrae sin reemplazo (generalmente es el caso), el tama침o de la muestra, n, no debe ser m치s del 10% de la poblaci칩n.

- **Suposici칩n de Tama침o de la Muestra**: Ahora, 쯖칩mo determinaremos qu칠 tan grande debe ser este tama침o? Bueno, depende de la poblaci칩n. Cuando la poblaci칩n est치 sesgada o es asim칠trica, el tama침o de la muestra debe ser grande. Si la poblaci칩n es sim칠trica, entonces podemos tomar muestras peque침as tambi칠n.

En general, un tama침o de muestra de 30 se considera suficiente cuando la poblaci칩n es sim칠trica.

### Ejemplo:

Validaremos el teorema del l칤mite central utilizando una base de datos real.

Consideremos la base de datos Used Cars Dataset (un conjunto de datos que contiene anuncios de autom칩viles con numerosas caracter칤sticas categ칩ricas y num칠ricas). En particular,  nos enfocaremos en la variable aleatoria usd_price.  A continuaci칩n se detallan los pasos a seguir: 

**Paso 1:** Escojamos inicialmente como n=5 el tama침o de la muestra aleatoria de la variable price_usd. 

**Paso 2:** Realizaremos 1000 muestras aleatorias con un tama침o de muestra de 5 de la columna price_usd y calcularemos el promedio de cada muestra.

**Paso 3:** : Usaremos la biblioteca Matplotlib para generar un histograma que ilustre la distribuci칩n de los 1000 promedios de muestras de tama침o 5, estableciendo el par치metro bins en 30. 

Verificaremos que la distribuci칩n de las medias de estas muestras tiende a seguir una distribuci칩n normal. Esta aproximaci칩n se volver치 m치s precisa a medida que aumentemos el tama침o de las muestras, por lo que repetiremos el c치lculo tomando muestras de tama침o 10 y 30. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
 

# Cargamos la base de datos cars.csv, previamente guardada en la carpeta de trabajo:
#https://www.kaggle.com/datasets/austinreese/craigslist-carstrucks-data
archivo_csv = 'cars.csv'

# Lee el archivo CSV
data_frame = pd.read_csv(archivo_csv)

# Columna que deseamos leer
nombre_columna = 'price_usd'

# Accedemos a la columna espec칤fica
columna_seleccionada = data_frame[nombre_columna]
media_price =np.mean(columna_seleccionada)
# Reemplaza 'tama침o_muestra' con el tama침o deseado para cada muestra
tama침o_muestra = 20

# Creamos las 1000 muestras aleatorias, cada una de tama침o tama침o_muestra, de la columna price_usd
muestras = [columna_seleccionada.sample(tama침o_muestra) for _ in range(1000)]

# Calcula el promedio de cada muestra
promedios = [muestra.mean() for muestra in muestras]

# Crea un histograma del promedio de las muestras
plt.hist(promedios, bins=30, edgecolor='black')  # Puedes ajustar el n칰mero de bins con otros valores
plt.title(f'Histograma del Promedio de Muestras price_usd (n={tama침o_muestra})')
plt.xlabel('Promedio de Muestra' )
plt.ylabel('Frecuencia')
plt.axvline(x=media_price,color='r') # for giving mean line
plt.show()


```

### Ejemplo:

Las latas de comida se envasa en vastas l칤neas de ensamblaje. Un tama침o particular de lata de sopa debe tener 19oz, pero hay variabilidad de una lata a otra  es una variable aleatoria. Supongamos que la distribuci칩n del contenido es aproximadamente normal con una media de 19.1 oz y una desviaci칩n est치ndar de 0.05.

a) Encuentra la probabilidad de que una lata seleccionada al azar contenga menos de 19oz. 

b) Los almacenes mayoristas venden estas latas en paquetes de cuatro. Encuentra la probabilidad de que el promedio de un paquete seleccionado al azar de cuatro latas sea inferior a 19oz.  

c) 쮼ntre qu칠 par de valores encontrar칤as el contenido promedio del 95% central de los paquetes de cuatro latas?

d) 쯇or qu칠 podr칤amos usar la distribuci칩n normal para encontrar las probabilidades en este problema?

### Soluci칩n:

a) En este caso una lata de comida tiene una distribuci칩n normal con media $\mu=19.1$ y desviaci칩n est치ndar de $0.05$. Por tanto la probabilidad de que una lata seleccionada al azar contenga menos de 19 oz es

$$P(X<19)= 0.022$$

Es decir, la probabilidad de que una lata seleccionada al azar contenga menos de 19 oz es $2,3\%$

b) En este caso, podemos aplicar el teorema del l칤mite central que afirma que si $\bar{X}_4$ denota el contenido promedio de una muestra aleatoria de 4 latas entonces $\bar{X}_4$ tiene distribuci칩n normal con media $\mu = 19.1$ y desviaci칩n est치ndar 
$$\sigma = \frac{0.05}{\sqrt{4}}$$.

Por tanto $$P(\bar{X}_4<19)=0.000032$$

La probabilidad de que el promedio de un paquete seleccionado al azar de cuatro latas sea inferior a 19oz es $0.0032\%$ 

c) Podemos usar la regla empirica de la desviaci칩n est치ndar que afirma que el 95% de los datos se encuentran alrededor de la media una ditancia de dos desviaciones est치ndar:

<img src="normaldistribution.png">


Por tanto, dado que la desviaci칩n est치ndar de $\bar{X}_4$ es  $\sigma =0.025 $ el par de valores donde se encontrar치 el contenido promedio del 95% central de los paquetes de cuatro latas es

$$19.1- 2*0.025 = 19.05$$
$$19.1+ 2*0.025 = 19.15$$

d) Podemos usar la distribuci칩n normal dado que el teorema del l칤mite central permite concluir que $\bar{X}_4$ tiene una distribuci칩n normal dado que $X$ tiene una distribuci칩n normal a pesar de que el tama침o de la muestra es peque침o.

```python
# C치lculos necesarios para resolver el problema anterior.

#importando el m칩dulo scipy.stats, que proporciona una variedad de funciones y distribuciones estad칤sticas.
import scipy.stats as stats

mean = 19.1
dev = 0.05
# a) 
x= 19
prob_less_than19= stats.norm.cdf( x, mean, dev)
print(f'P(X<19)={prob_less_than19:.3f}')

#b) 
mean = 19.1
dev = 0.05/(4)**(0.5)
x= 19

prob_less_than19= stats.norm.cdf( x, loc= mean, scale = dev)
print(f'P(X_4 <19)={prob_less_than19:.6f}')

# c) 
print(dev)
```

### Ejemplo: 

Supongamos que **X** representa el n칰mero de pacientes con ataques card칤acos que llegan a una cl칤nica por semana. La distribuci칩n de **X** tiene una media de 3.5 pacientes y una desviaci칩n est치ndar de 1.7.

a) 쯇uede la variable aleatoria **X** tener una distribuci칩n binomial?
   - No, la variable aleatoria $X$ no tiene las caracteristicas de una distribuci칩n binomial. Por ejemplo, no hay un n칰mero de ensayos y tampoco una probabilidad de 칠xito.

b) 쯇uede la variable aleatoria **X** tener una distribuci칩n normal?
   - No, la distribuci칩n de X debe estar sesgada hacia la derecha.

c) Supongamos que estamos interesados en $\bar{X}$, que es el n칰mero promedio de ataques card칤acos por semana en un a침o. 쮺u치l ser칤a el valor de **n**?
   - **n = 52**. Este valor corresponde al n칰mero de semanas en un a침o.

d) Encuentra la distribuci칩n de la media muestral.
   - Dado que **n = 52** es lo suficientemente grande, $\bar{X}$ sigue una distribuci칩n normal con media de 3.5 y desviaci칩n est치ndar de:
     $$\frac{1.7}{\sqrt{52}} = 0.236$$

e) Encuentra la probabilidad de que, en un a침o, haya un promedio de menos de 3 ataques card칤acos por semana en la cl칤nica.
   - Respuesta:
   Dado que $\bar{X}$ sigue aproximandamente una distribuci칩n normal, con media de 3.5 y desviaci칩n est치ndar de 0.236, obtenemos que :
     $$P(\bar{X} < 3) = P\left(Z < \frac{3 - 3.5}{0.236}\right) = P(Z < -2.12) = 0.0170$$
     Aproximadamente el 2% de los a침os tendr치n un promedio de menos de 3 ataques card칤acos.

### Ejemplo: 
Hay 40 estudiantes en una clase de estad칤stica elemental. Basado en a침os de experiencia, el instructor sabe que el tiempo requerido para calificar un primer examen seleccionado al azar es una variable aleatoria con un valor esperado de 6 min y una desviaci칩n est치ndar de 6 min. Si los tiempos de calificaci칩n son independientes y el instructor comienza a calificar a las 6:50 p.m., y califica en forma continua, 

a) 쯖u치l es la probabilidad (aproximada) de que termine de calificar antes de que se inicie el programa de noticias de las 11:00 p.m.?  

b)Si el reporte de deportes se inicia a la 11:10, 쯖u치l es la probabilidad de que se pierda una parte del reporte si espera
hasta que termine de calificar para prender la TV?
(Tomado de Devore, Probabilidad p. 229).



```python
# C치lculos necesarios para resolver el problema anterior.

#importando el m칩dulo scipy.stats, que proporciona una variedad de funciones y distribuciones estad칤sticas.
import scipy.stats as stats

mean = 6
dev = 6/(40**0.5)
#a) 
#Notemos que desde las 6:50 hasta las 11:00 tenemos 250 minutos. Por estudiante se tiene 250/40=6.25

x= 6.25
prob_less_than= stats.norm.cdf( x, mean, dev)
print(f'P(X<6.25)={prob_less_than:.4}')

#b)
#Notemos que desde las 6:50 hasta las 11:10 tenemos 260 minutos. Por estudiante se tiene 260/40=6.5
x = 6.5
prob_greather_than= 1- stats.norm.cdf( x, mean, dev)
print(f'P(X>6.5)={prob_greather_than:.4}')
```
