```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# DATA IMPORT AND FREQUENCY CALCULATION
data = pd.read_csv('StudentsPerformance.csv')
race_counts = data['race/ethnicity'].value_counts()


N = race_counts.sum()

# DEFINE THE DISCRETE RANDOM VARIABLE X MAPPING
# X = 1: group C
# X = 2: group D
# X = 3: group B
# X = 4: others (Group A + Group E)

# Create a mapping for the categories to their X values
x_map = {
    'group C': 1,
    'group D': 2,
    'group B': 3,
    'group A': 4,
    'group E': 4
}

# Group the counts based on the X value
grouped_counts = race_counts.groupby(race_counts.index.map(x_map)).sum()
grouped_counts.index.name = 'X_value (x)'

# Create a DataFrame for the distribution
df_dist = grouped_counts.reset_index(name='Absolute Frequency (f)')

# Map X values back to their group names for clarity
group_names = {1: 'Group C', 2: 'Group D', 3: 'Group B', 4: 'Others (A+E)'}
df_dist['Race/Ethnicity Group'] = df_dist['X_value (x)'].map(group_names)


# BUILD THE PROBABILITY DISTRIBUTION TABLE
# Calculate Relative Frequency (Probability) P(X=x)
df_dist['Probability P(X=x)'] = df_dist['Absolute Frequency (f)'] / N

print("Probability Distribution Table of X")
print("---------------------------------------")
print(f"Total Observations (N): {N}")
print(df_dist[['X_value (x)', 'Race/Ethnicity Group', 'Absolute Frequency (f)', 'Probability P(X=x)']].to_string(index=False))

# VERIFY THAT THE SUM OF PROBABILITIES EQUALS 1
sum_of_probabilities = df_dist['Probability P(X=x)'].sum()
print("\nProbability Sum Verification")
print("---------------------------------")
print(f"Sum of Probabilities: {sum_of_probabilities:.3f}")
if np.isclose(sum_of_probabilities, 1.0):
    print("Verification: The sum is equal to 1.0, confirming a valid probability distribution.")
else:
    print("Verification: The sum is NOT equal to 1.0.")
```

```python
# GRAPHICAL REPRESENTATION (BAR CHART)
plt.figure(figsize=(8, 5))
plt.bar(df_dist['X_value (x)'], df_dist['Probability P(X=x)'], 
        color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
plt.title('Probability Distribution of Race/Ethnicity (X)')
plt.xlabel('X Value (1:Group C, 2:Group D, 3:Group B, 4:Others)')
plt.ylabel('Probability P(X=x)')
plt.xticks(df_dist['X_value (x)'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# CALCULATE EXPECTED VALUE E(X) AND VARIANCE Var(X)

# Calculate x * P(X=x)
df_dist['x * P(X=x)'] = df_dist['X_value (x)'] * df_dist['Probability P(X=x)']

# Calculate x^2 * P(X=x)
df_dist['x^2 * P(X=x)'] = (df_dist['X_value (x)']**2) * df_dist['Probability P(X=x)']

# Expected Value E(X)
E_X = df_dist['x * P(X=x)'].sum()

# Variance Var(X)
E_X_squared = df_dist['x^2 * P(X=x)'].sum()
Var_X = E_X_squared - (E_X**2)
Std_Dev_X = np.sqrt(Var_X)

print("\nExpected Value and Variance")
print("---------------------------------")
print(f"Expected Value E(X): {E_X:.3f}")
print(f"Variance Var(X): {Var_X:.3f}")
print(f"Standard Deviation StdDev(X): {Std_Dev_X:.3f}")

# Explanation of E(X) and Var(X)
print("\nMeaning of the Values")
print(f"""Expected Value E(X) ≈ {E_X:.3f}: This is the weighted average of the encoded race/ethnicity 
      codes. It indicates the center of the distribution is numerically near 2.329, which is closest 
      to Group D (X=2). This does not represent an actual category but is a measure of central tendency 
      based on the numerical encoding.""")
print(f"""\nVariance Var(X) ≈ {Var_X:.3f}: This measures the dispersion of the encoded groups around 
      the mean. A value of 1.317 indicates a moderate spread, showing that the student population is 
      distributed significantly across all four coded categories, not heavily concentrated in just 
      one or two.""")
```

```python
# CALCULATE P(X <= 3) AND INTERPRET ---

# P(X <= 3) = P(X=1) + P(X=2) + P(X=3)
P_X_le_3 = df_dist[df_dist['X_value (x)'] <= 3]['Probability P(X=x)'].sum()

print("\nProbability P(X ≤ 3)")
print("---------------------------")
print(f"P(X ≤ 3) = P(X=1) + P(X=2) + P(X=3)")
print(f"P(X ≤ 3) = {P_X_le_3:.3f}")

# Interpretation
print("\nInterpretation of P(X ≤ 3)")
print(f"""The probability P(X ≤ 3) is 0.771 (or 77.1%). This is the probability 
      that a randomly selected student belongs to Group C, Group D, or Group B. 
      In the context of the study, this means that these three groups combined 
      represent over three-quarters (77.1%) of the students in the dataset.""")
```
