```python
import pandas as pd

df = pd.read_csv('StudentsPerformance.csv')

N_total = len(df)


# QUESTÃO 1: Probabilidade de que um estudante obtenha menos de 72 pontos em escrita ---
print("QUESTÃO 1: P(Escrita < 72)")

N_escr_menor_72 = df[df['writing score'] < 72].shape[0]

prob_escrita_menor_72 = N_escr_menor_72 / N_total

print(f"Total de Estudantes (N): {N_total}")
print(f"Estudantes com Escrita < 72: {N_escr_menor_72}")
print(f"Probabilidade P(W < 72): {prob_escrita_menor_72:.3f} ({prob_escrita_menor_72 * 100:.1f}%)")


# --- QUESTÃO 2: Probabilidade de que um estudante obtenha mais de 80 pontos em matemática E em escrita ---
print("\n--- QUESTÃO 2: P(Matemática > 80 E Escrita > 80) ---")

condicao_intersecao = (df['math score'] > 80) & (df['writing score'] > 80)
N_ambos_maior_80 = df[condicao_intersecao].shape[0]

prob_ambos_maior_80 = N_ambos_maior_80 / N_total

print(f"Total de Estudantes (N): {N_total}")
print(f"Estudantes com Matemática > 80 E Escrita > 80: {N_ambos_maior_80}")
print(f"Probabilidade P(M > 80 ∩ W > 80): {prob_ambos_maior_80:.3f} ({prob_ambos_maior_80 * 100:.1f}%)")


# --- QUESTÃO 3: Nível educacional dos pais com maior probabilidade condicional P(Matemática > 90 | Nível dos Pais) ---
# Evento A|B: P(Math Score > 90 | Parental Level)
print("\n--- QUESTÃO 3: Maior P(Matemática > 90 | Nível dos Pais) ---")

#Identificar o Evento A (Matemática > 90) e criar um DataFrame booleano para ele
evento_A = df['math score'] > 90

#Obter a lista de todos os níveis educacionais dos pais (Eventos B)
niveis_pais = df['parental level of education'].unique()

probabilidades_condicionais = {}

for nivel in niveis_pais:
    evento_B = df['parental level of education'] == nivel

    # Numerador: P(A ∩ B_i) -> Contagem de estudantes com M > 90 E o Nível B_i
    # Sum() em uma série booleana conta o número de True
    N_A_e_B = (evento_A & evento_B).sum()

    # Denominador: P(B_i) -> Contagem total de estudantes com o Nível B_i
    N_B = evento_B.sum()
    
    # Cálculo da Probabilidade Condicional: P(A | B_i) = N(A ∩ B_i) / N(B_i)
    if N_B > 0:
        prob = N_A_e_B / N_B
        probabilidades_condicionais[nivel] = prob
    else:
        probabilidades_condicionais[nivel] = 0

series_prob = pd.Series(probabilidades_condicionais).sort_values(ascending=False)

print("\nCálculo da Probabilidade Condicional por Nível:")
print(series_prob.to_string(float_format="%.3f"))

melhor_nivel = series_prob.index[0]
maior_prob = series_prob.iloc[0]

print("\nNível dos pais que fornece a MAIOR probabilidade condicional:")
print(f"-> {melhor_nivel} com P(M > 90 | Nível) = {maior_prob:.3f} ({maior_prob * 100:.1f}%)")
```
