```python
from scipy.stats import binom
```

```python
help(binom)
```

```python
# Número de ensayos( 20 preguntas)
n = 20
# Probabilidd de éxito en cada ensayo
p = 0.2
```

```python
#P(X=10)

prob_X_10 = binom.pmf(10,n,p)
```

```python
print(f' La probabilidad de obtener 10 respuestas correctas es {prob_X_10*100:.3f}%')
```

```python
#P(X>=2)=1-P(X<=1)


prob_X_mayor_o_igual_2 = 1- binom.cdf(1,20,0.2)
```

```python
prob_X_mayor_o_igual_2
```

```python
# También se puede calcular el valor anterior así:
prob_X_mayor_o_igual_2 = 1- (binom.pmf(0,n,p)+binom.pmf(1,n,p))
```

```python
prob_X_mayor_o_igual_2 
```

```python
print(f' La probabilidad de obtener al menos 2 respuestas correctas es {prob_X_mayor_o_igual_2*100:.3f}%')
```

 # Parte 3
 $$P(2\leq X \leq 9)= P(X\leq 9)-P(X\leq 1)$$

```python
# Número de ensayos( 20 preguntas)
n = 20
# Probabilidd de éxito en cada ensayo
p = 0.2


prob_x_2_9 = binom.cdf(9,n,p) - binom.cdf(1,n,p)
```

```python
print(prob_x_2_9)
```

```python
print(f' La probabilidad de obtener entre 2 y 9 respuestas correctas es {prob_x_2_9*100:.3f}%')
```
